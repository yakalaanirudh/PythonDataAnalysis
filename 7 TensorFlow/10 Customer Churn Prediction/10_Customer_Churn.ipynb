```python
"""
Customer churn prediction is to measure why customers are leaving a business. 
In this code we will be looking at customer churn in telecom business. 
We will build a deep learning model to predict the churn and use precision,recall, f1-score to measure performance of our model
"""
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv(r"D:customer_churn.csv")
df.sample(5)
```

```python
#we drop customerID as it has no value in machine learning
df.drop('customerID',axis='columns',inplace=True)
df.dtypes
```

```python
"""
We can see that the totalcharges is an object instead of float
convert those to numbers since all empty strings and other charcater give an error we use coerce
"""
df.TotalCharges.values
#pd.to_numeric(df.TotalCharges)
pd.to_numeric(df.TotalCharges,errors='coerce').isnull()
```

```python
#Here we select the dataframe where the value for totalcharges is null
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 

print_unique_col_values(df1)
```

```python
#Some of the columns have no internet service or no phone service, that can be replaced with a simple No

df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)

print_unique_col_values(df1)
```

```python
#Convert Yes and No to 1 or 0

yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)


for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
df1.gender.unique()
```

```python
#One hot encoding for categorical columns

df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns

df2.sample(5)
```

```python
df2.dtypes
```

```python
#Scaling
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
df2[[
'InternetService_DSL','InternetService_Fiber optic','InternetService_No',
'Contract_Month-to-month','Contract_One year','Contract_Two year',
'PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)',
'PaymentMethod_Electronic check','PaymentMethod_Mailed check']].replace({'True':1,'False':0},inplace=True)

for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
#Train test split

X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
#Building model using Artifical Neural Network ANN
"""
The first layer expects an input with 26 features (input_shape=(26,)).
The Dense(26, ...) layer has 26 neurons, each connected to all 26 input features.
"""
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)

```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
